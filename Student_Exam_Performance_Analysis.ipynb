### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Reading Dataset

```python
data = pd.read_csv('data/raw/Student_Performance_Factors (raw_data).csv')
data.head()
```

```python
df = data.copy()
```

### Check for Null Values

```python
df.info()
```

### Rearranging the columns

```python
data.columns.values
```

```python
col_names = [
    'Student_id',
    'Gender',
    'School_Type',
    'Parental_Involvement',
    'Parental_Education_Level',
    'Motivation_Level',
    'Teacher_Quality',
    'Peer_Influence',
    'Distance_from_Home',
    'Internet_Access',
    'Access_to_Resources',
    'Extracurricular_Activities',
    'Learning_Disabilities',
    'Family_Income',
    'Hours_Studied',
    'Attendance',
    'Sleep_Hours',
    'Physical_Activity',
    'Previous_Scores',
    'Tutoring_Sessions',
    'Exam_Score'
]

df = df[col_names]
df.head()
```

### Check duplicates

```python
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Confirm
df.duplicated().sum()
```

### Handling Missing Values

#### *Exam_Score*

```python
df['Exam_Score'].isnull().sum()
```

```python
df = df.dropna(subset=['Exam_Score'])
```

#### *Attendance*

```python
df['Attendance'].isnull().sum()
```

```python
df = df.dropna(subset=['Attendance'])
```

#### *Hours_Studied*

```python
df['Hours_Studied'].isnull().sum()
```

```python
df = df.dropna(subset=['Hours_Studied'])
```

#### *Previous_Scores*

```python
df['Previous_Scores'].isnull().sum()
```

```python
median_prev_score = df['Previous_Scores'].median()
df['Previous_Scores'] = df['Previous_Scores'].fillna(median_prev_score)
```

#### *Motivation_Level*

```python
df['Motivation_Level'].isnull().sum()
```

```python
df['Motivation_Level'] = df['Motivation_Level'].fillna("Unknown")
```

#### *Tutoring_Sessions*

```python
df['Tutoring_Sessions'].isnull().sum()
```

```python
df['Tutoring_Sessions'] = df['Tutoring_Sessions'].fillna(0)
```

#### *CREATE ENGINEERED COLUMN – Performance_Level*

```python
def performance_level(score):
    if score >= 75:
        return "Top Performer"
    elif score >= 50:
        return "Average Performer"
    else:
        return "Weak Performer"

df['Performance_Level'] = df['Exam_Score'].apply(performance_level)
```

```python
df['Score_per_Hour'] = df['Exam_Score'] / df['Hours_Studied']
```

```python
df[
    ['Exam_Score', 'Attendance', 'Hours_Studied',
     'Previous_Scores', 'Motivation_Level',
     'Tutoring_Sessions', 'Performance_Level']
].isnull().sum()
```
